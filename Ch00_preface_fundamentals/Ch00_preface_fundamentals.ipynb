## 1. 钱为什么会"变多"或"变少"？ / 1. Why Does Money Gain or Lose Value?

### 通货膨胀：钱在变"薄" / Inflation: Money Getting Thinner

> 10年前一碗拉面 15块，现在要 30块。面还是那碗面，钱还是那些钱，但钱"变薄了"。
>
> Ten years ago a bowl of ramen cost 15 yuan; today it costs 30. Same ramen, same banknote, but the banknote is "thinner".

这就是**通货膨胀** —— 物价上涨，同样的钱能买到的东西变少了。

That is **inflation** — prices rise, so the same amount of money buys fewer goods.

所以人们要想办法让钱"跑赢"通胀，这就产生了——

People therefore look for ways to make their money "outrun" inflation, and that gives rise to —

### 投资 vs 投机 vs 赌博 / Investing vs. Speculation vs. Gambling

|  | 投资 / Investing | 投机 / Speculation | 赌博 / Gambling |
|--|------|------|------|
| 目的 / Purpose | 长期增值 / Long-term wealth growth | 短期获利 / Short-term profit | 碰运气 / Try one's luck |
| 依据 / Basis | 分析、研究 / Analysis & research | 趋势、时机 / Trend & timing | 随机 / Random |
| 时间 / Horizon | 几年到几十年 / Years to decades | 几天到几个月 / Days to months | 几秒到几小时 / Seconds to hours |
| 例子 / Examples | 买房、定投基金 / Buying property, fund SIPs | 炒股、做外汇 / Stock trading, FX | 买彩票、猜大小 / Lotteries, dice games |

> **本教程的内容主要属于"投机"范畴** —— 通过分析价格走势来短期获利。
> 这不是坏事，但需要知识和纪律。没有知识的投机 = 赌博。
>
> **This tutorial mostly lives in the "speculation" bucket** — profiting short-term by analysing price action.
> That is not a bad thing, but it requires knowledge and discipline. Speculation without knowledge = gambling.

## 2. 金融市场里有什么？ / 2. What Is Sold in Financial Markets?

想象一个巨大的"超市"，里面卖的不是食品，而是各种"金融商品"：

Imagine a giant "supermarket" whose shelves hold financial products instead of food:

### 2.1 股票 — 公司的"碎片" / 2.1 Stocks — Slices of a Company

> 苹果公司值 3万亿美元。它把自己切成 150亿 份，每份就是一股"股票"。
> 你买了 1 股 = 你拥有苹果公司的 150亿分之一。
>
> Apple is worth USD 3 trillion. It cuts itself into 15 billion slices; each slice is one share.
> Owning 1 share = owning 1 / 15 billion of Apple.

- 公司赚钱 → 股价涨 → 你赚了 / Company earns money → share price rises → you profit
- 公司亏钱 → 股价跌 → 你亏了 / Company loses money → share price falls → you lose

### 2.2 债券 — 借条 / 2.2 Bonds — IOUs

> 政府或公司需要钱，向你借。
> 给你打一张"借条"（债券），承诺每年给你利息，到期还本金。
>
> A government or company needs cash and borrows it from you.
> It hands you an IOU (a bond), promising yearly interest and full principal at maturity.

- 比股票安全（利息是固定的）/ Safer than stocks (interest is fixed)
- 但收益也低 / But the return is smaller

### 2.3 外汇 — 不同国家的钱互相兑换 / 2.3 Foreign Exchange — Trading One Country's Money for Another

> 你去日本旅游，要把人民币换成日元。
> 汇率 = 1人民币 能换多少日元。
>
> Flying to Japan? You swap RMB for JPY.
> Exchange rate = how many JPY one RMB buys.

汇率每时每刻都在变。有人专门靠预测汇率变化来赚钱，这就是**外汇交易**。

Exchange rates move every second. Some people make a living predicting those moves — that is **FX trading**.

### 2.4 商品 — 实实在在的东西 / 2.4 Commodities — Tangible Goods

> 黄金、石油、小麦、铜... 这些实物也可以在金融市场上买卖。
>
> Gold, oil, wheat, copper... physical goods that also trade on financial markets.

### 2.5 加密货币 — 数字世界的"钱" / 2.5 Crypto — Digital-World "Money"

> 比特币、以太坊... 不由任何政府发行，靠技术保障安全。
> 波动极大——可能一天涨20%，也可能一天跌30%。
>
> Bitcoin, Ethereum... issued by no government, secured by technology.
> Extremely volatile — up 20 % in a day, or down 30 % in a day.

## 3. "价格"是怎么决定的？ / 3. How Is a Price Decided?

> **一句话：买的人多就涨，卖的人多就跌。**
>
> **One-liner: more buyers than sellers → price up; more sellers than buyers → price down.**

这叫做**供需关系**：

That is the law of **supply and demand**:

```
很多人想买 + 很少人想卖 → 供不应求 → 价格涨
很多人想卖 + 很少人想买 → 供过于求 → 价格跌

many buyers + few sellers  → demand > supply → price rises
many sellers + few buyers  → supply > demand → price falls
```

价格不停地变动，就形成了**行情**。交易者的工作就是判断价格接下来会涨还是跌。

Prices move non-stop, producing what traders call **market action**. A trader's job is to guess which way it goes next.

## 4. "买"和"卖" / 4. "Buy" and "Sell"

在普通生活中，你只能"先买后卖"：

In everyday life you can only buy first, sell later:

> 100块买了一个东西 → 涨到120块卖掉 → 赚了20块
>
> Buy something for 100 → sell it at 120 → pocket 20

在金融市场中，你还可以"先卖后买"（这叫**做空**）：

In financial markets you can also sell first, buy later — that is **going short**:

> 你觉得一个东西会跌，先以100块"借来卖掉" → 跌到80块再"买回来还" → 赚了20块
>
> Think something will fall → "borrow and sell" at 100 → "buy back and return" at 80 → pocket 20

| 术语 / Term | 意思 / Meaning | 什么时候赚钱 / When You Profit |
|------|------|-----------|
| **做多 (Long / Buy)** | 先买后卖 / Buy first, sell later | 价格**涨了**赚钱 / Price **rises** |
| **做空 (Short / Sell)** | 先卖后买 / Sell first, buy later | 价格**跌了**赚钱 / Price **falls** |

> 做空听起来奇怪（怎么能卖自己没有的东西？），但在金融市场中这是完全合法和常见的操作。
> 平台会帮你处理"借"的过程。
>
> Shorting sounds strange (how can you sell what you do not own?), but it is fully legal and extremely common. Your broker handles the "borrowing" for you behind the scenes.

## 5. 什么是"利率"？ / 5. What Is an "Interest Rate"?

> **利率 = 借钱的代价（或存钱的回报）**
>
> **Interest rate = the cost of borrowing money (or the reward for saving it)**

你把钱存银行，银行给你利息（年利率3%=每存100块，一年后多3块）。

Deposit cash in a bank → the bank pays you interest (3 % annual = deposit 100, get 3 extra after a year).

你找银行贷款，你要给银行利息（房贷利率5%=每借100万，一年要付5万利息）。

Take a loan → you pay the bank interest (5 % mortgage rate = borrow 1 000 000, pay 50 000 interest per year).

**利率为什么重要？** 因为它影响一切：

**Why do rates matter?** Because they touch everything:

| 利率上升时 / Rates Going Up | 利率下降时 / Rates Going Down |
|-----------|-----------|
| 借钱变贵 → 人们少借钱 → 消费减少 / Borrowing gets pricey → less borrowing → less spending | 借钱变便宜 → 人们多借钱 → 消费增加 / Borrowing gets cheap → more borrowing → more spending |
| 存钱更划算 → 不想投资 → 股市倾向下跌 / Saving looks better → less investing → stocks tend down | 存钱没意思 → 想去投资 → 股市倾向上涨 / Saving pays poorly → more investing → stocks tend up |
| 本国货币变强（外国人存钱来了）/ Home currency strengthens (foreigners park cash here) | 本国货币变弱 / Home currency weakens |

各国的**中央银行**通过调整利率来控制经济。这对外汇和所有金融市场都有巨大影响。

**Central banks** steer their economies by moving interest rates. This has a massive impact on FX and every other financial market.

## 6. 什么是"风险"？ / 6. What Is "Risk"?

> **风险 = 你实际拿到的结果和预期不一样的可能性。**
>
> **Risk = the chance that what you actually get differs from what you expected.**

- 你买了股票以为会涨，结果跌了 → 这就是风险 / Bought a stock expecting it to rise, it fell → that is risk
- 你存银行很安全，但回报也低 → 低风险低回报 / Bank deposits are safe, but pay little → low risk, low return

### 风险和回报的关系 / The Risk – Return Trade-off

```
高风险 ←————————————————→ 低风险
High risk ←——————————————→ Low risk
高回报                     低回报
High return                 Low return

加密货币  股票  基金  债券  银行存款
Crypto    Stocks  Funds  Bonds  Bank deposits
  ↑                           ↑
波动大                      波动小
High volatility           Low volatility
可能暴富                    稳定但慢
Can make you rich          Steady but slow
也可能归零
Can also go to zero
```

> **核心原则：天下没有免费的午餐**
> 想赚更多 → 必须承担更大的风险
> 不想承受风险 → 回报就很有限
>
> **Golden rule: there is no such thing as a free lunch.**
> Want higher returns → accept higher risk.
> Refuse to take risk → your upside is tiny.

## 7. 互动体验：看看"价格波动"是什么感觉 / 7. Interactive Demo: What Does "Price Volatility" Feel Like?

> 下面模拟一个金融产品的价格在一年内的随机波动。
> 拖动**波动率**滑块，感受低波动和高波动的区别。
>
> The cell below simulates one year of random price movement for a financial product.
> Drag the **volatility** slider to feel the difference between calm and wild markets.

````python
def price_volatility_demo(start_price=100, volatility=20, n_paths=5):
    """Simulate random price paths to visualize different volatility levels."""
    plt.close('all')
    fig, axes = plt.subplots(1, 2, figsize=(13, 5))
    days = 252  # one trading year
    t = np.arange(days)

    # --- Left: price paths ---
    ax = axes[0]
    final_values = []
    for i in range(n_paths):
        np.random.seed(42 + i)
        daily_returns = np.random.normal(0.0002, volatility / 100 / np.sqrt(252), days)
        path = start_price * np.cumprod(1 + daily_returns)
        color = plt.cm.tab10(i / n_paths)
        ax.plot(t, path, lw=1.5, alpha=0.7, color=color)
        final_values.append(path[-1])
    ax.axhline(y=start_price, color='gray', ls='--', alpha=0.5, label=f'Start {start_price}')
    ax.set_xlabel('Day (1 year = 252 trading days)', fontsize=11)
    ax.set_ylabel('Price', fontsize=11)
    ax.set_title(f'Price Paths (Volatility = {volatility}%)', fontsize=13, fontweight='bold')
    ax.legend(fontsize=9)
    ax.grid(True, alpha=0.2)

    # --- Right: terminal distribution ---
    ax = axes[1]
    np.random.seed(0)
    all_ends = []
    for i in range(1000):
        daily = np.random.normal(0.0002, volatility / 100 / np.sqrt(252), days)
        all_ends.append(start_price * np.prod(1 + daily))
    all_ends = np.array(all_ends)

    ax.hist(all_ends, bins=50, color=C_BLUE, alpha=0.6, edgecolor='white')
    ax.axvline(x=start_price, color=C_RED, ls='--', lw=2, label=f'Start {start_price}')
    ax.axvline(x=np.mean(all_ends), color=C_GREEN, ls='--', lw=2, label=f'Mean end {np.mean(all_ends):.1f}')
    win_pct = np.mean(all_ends > start_price) * 100
    ax.set_xlabel('Price after 1 year', fontsize=11)
    ax.set_ylabel('Count', fontsize=11)
    ax.set_title(f'1000-path Terminal Distribution ({win_pct:.0f}% prob up)', fontsize=12, fontweight='bold')
    ax.legend(fontsize=9)
    ax.grid(True, alpha=0.2)

    plt.tight_layout()
    plt.show()

    v_max = max(all_ends)
    v_min = min(all_ends)
    print(f'  Volatility {volatility}% effect (1000 simulations):')
    print(f'    Best:  ${v_max:.0f} (+{(v_max / start_price - 1) * 100:.0f}%)')
    print(f'    Worst: ${v_min:.0f} (-{(1 - v_min / start_price) * 100:.0f}%)')
    print(f'    Prob up: {win_pct:.0f}%')
    if volatility > 40:
        print('  !! High volatility = big upside but big downside too. That is risk!')

interact(price_volatility_demo,
         start_price=IntSlider(value=100, min=10, max=1000, step=10, description='Start price:'),
         volatility=IntSlider(value=20, min=5, max=80, step=5, description='Volatility(%):'),
         n_paths=IntSlider(value=5, min=1, max=20, step=1, description='# Paths:'));
````

## 8. 金融市场的主要参与者 / 8. Who Actually Trades in the Markets?

| 角色 / Role | 他们是谁 / Who They Are | 他们在做什么 / What They Do |
|------|---------|------------|
| **散户 / Retail traders** | 普通人（你和我）/ Ordinary people (you and me) | 用自己的钱交易，资金量小 / Trade personal money in small size |
| **机构投资者 / Institutional investors** | 基金公司、保险公司、银行 / Fund companies, insurers, banks | 管理大量资金，影响市场 / Run huge pools of money, move markets |
| **中央银行 / Central banks** | 各国央行（美联储、欧央行）/ National central banks (Fed, ECB, ...) | 制定利率政策，影响整个经济 / Set rate policy, steer the whole economy |
| **做市商 / Market makers** | 专门提供买卖报价的机构 / Firms that always quote bid / ask | 保持市场流动性 / Keep markets liquid |
| **经纪商 (Broker)** | 连接你和市场的平台 / Platforms connecting you to the market | 你下单给他，他帮你成交 / You send orders, they route them to execution |
